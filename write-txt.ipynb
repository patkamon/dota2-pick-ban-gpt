```python
import pandas as pd
```

```python
df = pd.read_csv("pick-ban.csv")
```

```python
li = sorted(df['r_pick4'].unique().tolist())
```

```python
import json
dic = {}
for i, hero in enumerate(li):
    dic[str(i)] = hero 

with open("hero.json", "w") as f:
    json.dump(dic, f)
```

```python
# base on this order

# r ban1 - d ban1 - d ban2 - r ban2 - d ban3 - d ban4 - r ban3 - r pick1 - d pick1 - r ban4 - r ban5 - d ban5
# d pick2 - r pick2 - r pick3 - d pick3 - d pick4 - r pick4 - r ban6 - d ban6 - d ban7 - r ban7 - r pick5 - d pick5



with open('pick-ban.txt', 'a') as the_file:
    for i in range(len(df)):
        lin = ""
        focus = df.iloc[i]
        # order pick ban
        lin += focus["r_ban1"] + "/"
        lin += focus["d_ban1"] + "/"
        lin += focus["d_ban2"] + "/"
        lin += focus["r_ban2"] + "/"
        lin += focus["d_ban3"] + "/"
        lin += focus["d_ban4"] + "/"
        lin += focus["r_ban3"] + "/"
        lin += focus["r_pick1"] + "/"
        lin += focus["d_pick1"] + "/"
        lin += focus["r_ban4"] + "/"
        lin += focus["r_ban5"] + "/"
        lin += focus["d_ban5"] + "/"
        lin += focus["d_pick2"] + "/"
        lin += focus["r_pick2"] + "/"
        lin += focus["r_pick3"] + "/"
        lin += focus["d_pick3"] + "/"
        lin += focus["d_pick4"] + "/"
        lin += focus["r_pick4"] + "/"
        lin += focus["r_ban6"] + "/"
        lin += focus["d_ban6"] + "/"
        lin += focus["d_ban7"] + "/"
        lin += focus["r_ban7"] + "/"
        lin += focus["r_pick5"] + "/"
        lin += focus["d_pick5"]

        the_file.write(f'{lin}\n')
```
